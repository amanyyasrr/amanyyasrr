```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# تحميل MobileNetV2 بدون الطبقة الأخيرة
base_model = MobileNetV2(weights='imagenet', include_top=False)

# تجميد الطبقات الأساسية
base_model.trainable = False

# إضافة طبقات التصنيف المخصصة
x = base_model.output
x = GlobalAveragePooling2D()(x)  # تحويل الميزات إلى مصفوفة صغيرة
x = Dense(128, activation='relu')(x)  # طبقة مخفية
x = Dense(64, activation='relu')(x)  # طبقة أخرى
predictions = Dense(1, activation='sigmoid')(x)  # طبقة الإخراج (حقيقية أو مزيفة)

# إنشاء النموذج النهائي
model = Model(inputs=base_model.input, outputs=predictions)

# تجميع النموذج
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# عرض ملخص النموذج
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# مسار البيانات داخل Google Drive
data_dir = "/content/drive/MyDrive/R\F"

# تجهيز البيانات باستخدام Augmentation لتحسين دقة النموذج
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # تحويل القيم إلى نطاق [0,1]
    rotation_range=20,  # تدوير بسيط للصورة
    width_shift_range=0.2,  # إزاحة أفقية
    height_shift_range=0.2,  # إزاحة عمودية
    shear_range=0.2,  # تشويه بسيط للصورة
    zoom_range=0.2,  # تقريب / تكبير
    horizontal_flip=True,  # عكس الصور أفقيًا
    validation_split=0.2  # تخصيص 20% من البيانات للاختبار
)

# تحميل بيانات التدريب والاختبار
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # حجم الصورة المطلوب
    batch_size=32,
    class_mode='binary',  # لأن لدينا فئتين فقط (حقيقي/مزيف)
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# تدريب النموذج
history = model.fit(
    train_generator,
    epochs=10,  # عدد التكرارات
    validation_data=val_generator
)

# حفظ النموذج بعد التدريب
model.save("/content/drive/MyDrive/money_detector_model.h5")

print("✅ تم حفظ النموذج بنجاح!")
```

```python
import tensorflow as tf
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image


model = tf.keras.models.load_model("/content/drive/MyDrive/money_detector_model.h5")


image_path = "/content/drive/MyDrive/R\F/Real/1-american-dollar-banknote-obverse-1.jpg"


img = image.load_img(image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0
prediction = model.predict(img_array)


if prediction[0][0] > 0.5:
    print("🚨 Fake Money!")
else:
    print("✅ Real Dollar!")
```
