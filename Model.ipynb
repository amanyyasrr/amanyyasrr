```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt


base_model = MobileNetV2(weights='imagenet', include_top=False)


base_model.trainable = False


x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=predictions)

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator


data_dir = "/content/drive/MyDrive/R\F"

train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)


train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)


history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)


model.save("/content/drive/MyDrive/money_detector_model.h5")

```

```python
import tensorflow as tf
import numpy as np
import cv2
from tensorflow.keras.preprocessing import image


model = tf.keras.models.load_model("/content/drive/MyDrive/money_detector_model.h5")


image_path = "/content/drive/MyDrive/R\F/Fake/93-fake-dollar-bill-printable-fake-money-printable-australian-free-printable-fake-money-that-looks-real.jpg"


img = image.load_img(image_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0
prediction = model.predict(img_array)


if prediction[0][0] > 0.5:
    print("🚨 Fake Money!")
else:
    print("✅ Real Dollar!")
```
